```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
resDirecto = pd.read_csv("resDirecto.csv", header=None)
resJacobi = pd.read_csv("resJacobi.csv", header=None)
resGauss = pd.read_csv("resGauss.csv", header=None)
```

```python
def f(x,y):
    if(y==0 or y==1):
        return 0
    if(x==0 or x==1):
        return y*(1-y)
    np.nan
```

```python
aux = pd.DataFrame()
```

```python
for x in range(11):
    lista = []
    for y in range(11):
        lista.append(f(x/10,y/10))
    aux[x] = lista
```

```python
for col in range(10,-1,-1):
    resDirecto = resDirecto.rename(index = int, columns={col: col+1})
    resJacobi = resJacobi.rename(index = int, columns={col: col+1})
    resGauss = resGauss.rename(index = int, columns={col: col+1})
```

```python
for i in range(9):
    aux.loc[i+1,1:9] = resDirecto.loc[i,:]
resDirecto = aux.copy()
```

```python
for i in range(9):
    aux.loc[i+1,1:9] = resJacobi.loc[i,:]
resJacobi = aux.copy()
```

```python
for i in range(9):
    aux.loc[i+1,1:9] = resGauss.loc[i,:]
resGauss = aux.copy()
```

```python
%matplotlib notebook
fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(resDirecto.index, resDirecto.columns, resDirecto.values, cmap="viridis", vmin=0, vmax=0.3)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel('Tension')
fig.colorbar(surf)
ax.view_init(45, 225)
plt.draw()
```
