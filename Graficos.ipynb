```python
def f(x,y):
    if(y==0 or y==1):
        return 0
    if(x==0 or x==1):
        return y*(1-y)
    np.nan
```

```python
aux = pd.DataFrame()
```

```python
for x in range(11):
    lista = []
    for y in range(11):
        lista.append(f(x/10,y/10))
    aux[x] = lista
```

```python
for col in range(10,-1,-1):
    resDirecto = resDirecto.rename(index = int, columns={col: col+1})
    resJacobi = resJacobi.rename(index = int, columns={col: col+1})
    resGauss = resGauss.rename(index = int, columns={col: col+1})
```

```python
for i in range(9):
    aux.loc[i+1,1:9] = resDirecto.loc[i,:]
resDirecto = aux.copy()
```

```python
for i in range(9):
    aux.loc[i+1,1:9] = resJacobi.loc[i,:]
resJacobi = aux.copy()
```

```python
for i in range(9):
    aux.loc[i+1,1:9] = resGauss.loc[i,:]
resGauss = aux.copy()
```

```python
lista = []
for i in range(11):
    lista.append([])
    for j in range(11):
        lista[i].append(j)
x = np.array(lista)
y = x.transpose()
```

```python
y
```

```python
%matplotlib widget
from ipywidgets import HBox, FloatSlider
plt.ioff()
plt.clf()

slider = FloatSlider(orientation="vertical",
                    value=1.0,
                    min=0,
                    max=360
)

slider2 = FloatSlider(orientation="vertical",
                    value=1.0,
                    min=0,
                    max=90
)

fig = plt.figure(figsize=(8,8), num="Test")
fig.suptitle("Resultado 3D por Metodo Directo")
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(x, y, resDirecto.values, cmap="viridis", vmin=0, vmax=0.3)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel('Tension')
fig.colorbar(surf)
ax.view_init(slider2.value, slider.value)


def update_angle(change):
    ax.view_init(slider2.value, change.new)
    fig.canvas.draw()
    fig.canvas.flush_events()

def update_vertical(change):
    ax.view_init(change.new, slider.value)
    fig.canvas.draw()
    fig.canvas.flush_events()
    
slider.observe(update_angle, names="value")
slider2.observe(update_vertical, names="value")

HBox([slider, slider2, fig.canvas])
```
